```python
import pandas as pd
```

```python
df = pd.read_csv('../GlobalLandTemperaturesByCity.csv')
df.head()
```

```python
del df['AverageTemperatureUncertainty']
df['Year'] = df['dt'].map(lambda dt: dt.split('-')[0])
df['Month'] = df['dt'].map(lambda dt: dt.split('-')[1])
del df['dt']
df.head()
```

```python
df['New'] = df['Year']
df['New'] = df['New'].map(lambda year: str(year))
df['New'] = df['New'].map(lambda year: year + ", ")
df['AverageTemperature'] = df['AverageTemperature'].map(lambda temperature: str(temperature))
df['New'] = df['New'] + df['AverageTemperature']
df['New'] = df['New'] + ", " + df['City']
df = df[df.Year != '1743']
df = df[df.Year != '2013']
df = df.sort_values(['Year', 'City'], ascending=[1, 1])
```
